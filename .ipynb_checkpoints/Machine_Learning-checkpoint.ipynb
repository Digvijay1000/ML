## Machine Learning

**Machine learning is the branch of artifical intelliance (AI) and computer science which focuses on the use of data and algorithms to imitate(to follow or try) the way that humans learn, gradually improving its accuracy.**

    The reagion behind of ML to take insights from data with the help of statistical methods and algorithms. These insights subsequently drive decision making within applications and businesses.

Learning is any process by which a system improves performance from experience.

    Machine Learning is the study of algorithms that 
        
        • improve their performance P 
        • at some task T 
        • with experience E. 

    A well-defined learning task is given by <P, T, E>.


1. Optimize a performance criterion using example data or past experience.

2. Role of Statistics: Inference from a sample.

3. Role of Computer science: Efficient algorithms to Solve the optimization problem.

4. Representing and evaluating the model for inference (conclusion or opinion).

**Types of Learning**

• **Supervised (inductive) learning**

– Given: training data + desired outputs (labels)

• **Unsupervised learning**

– Given: training data (without desired outputs)

• **Semi-supervised learning**

– Given: training data + a few desired outputs

• **Reinforcement learning**

– Rewards from sequence of actions
